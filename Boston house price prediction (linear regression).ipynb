```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
boston_data = load_boston() #storing boston housing dataset in 'df' variable
```

```python
boston_data
```

```python
#since df is not organised, ill convert to to a proper dataframe using 'pd.dataframe()' function.
# '.data' has been typed after 'boston_data' because only the 'data' variable form 'boston_data' will get converted to dataframe
#check the below pic 
dataset = pd.DataFrame(boston_data.data) 
```

```python
dataset
```

```python
#setting column names
dataset.columns = boston_data.feature_names
```

```python
dataset.head()
```

```python
# setting Independent and dependent features/variables
X = dataset #independent variables
y = boston_data.target #dependent variables
```

```python
y
```

```python
## train test split 
from sklearn.model_selection import train_test_split

#X_train & y_train is used to train the model. X_test & y_test is used to test the model.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
## standardizing the dataset. this is done before applying linear regression
'''Standardization is useful when your data has varying scales such as linear regression, logistic regression, and 
linear discriminant analysis. Data standardization helps improve the quality of your data by transforming and 
standardizing it.
Data standardization is about making sure that data is internally consistent; that is, each data type has the same content 
and format.
Standardization comes into the picture when features of the input data set have large differences between their ranges, or 
simply when they are measured in different units (e.g., pounds, meters, miles, etc.).
These differences in the ranges of initial features cause trouble for many machine learning models. 
For example, for the models that are based on distance computation, if one of the features has a broad range of values, 
the distance will be governed by this particular feature.'''

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() # initializing standard scalar
```

```python
#i will standardize only the independent/input variables of the train & test data.
X_train = scaler.fit_transform(X_train)

#scaler.inverse_transform(X_train) -> to reverse the transform and get the actual dataset.
```

```python
X_test = scaler.transform(X_test)
```

```python
'''
The fit(data) method is used to compute the mean and std dev for a given feature to be used further for scaling. 
The transform(data) method is used to perform scaling using mean and std dev calculated using the .fit() method. 
The fit_transform() method does both fits and transform
'''
```

```python
#now ill build the linear regression model.
#this is a multiple linear regession problem because we have two or more independent variables(X) and 
#one dependent variable(y).
```

```python
from sklearn.linear_model import LinearRegression

#for cross validation. 
'''in Cross-Validation the training data is internally divided into multiple train test splits for validating the 
model efficiency. each cross validation has a different accuracy.
the mean of the combined accuracies will be our final output'''

from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse = cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=10)

#The Mean Squared Error measures how close a regression line is to a set of data points.Mean square error is calculated 
#by taking the average, specifically the mean, of errors squared from data as it relates to a function.
#mse is the difference between predicted & actual values
#the lower the value of mse the better and 0 means the model is perfect.
#cv = 10 because i am cross validating 10 times.
```

```python
mse
```

```python
np.mean(mse) #calculatin the mean of mse
```

```python
#prediction of test data
reg_pred = regression.predict(X_test) #the predicted values will be stored in reg_pred
```

```python
reg_pred #below are the predicted values
```

```python
import seaborn as sns

#i will check the distance the between the predicted values(reg_pred) & the actual values(y_test)
sns.displot(reg_pred-y_test,kind='kde') #distplot is distance plot

#from the belwo plot i can see that the distance between predicted values(reg_pred) & the actual values(y_test) is less
#most of the variation between predicted values(reg_pred) & the actual values(y_test) is between -10 & 10.
```

```python
#computing r2 score(r square score) aka Goodness of fit. R2 score is a metric that tells the performance of your model.
#as our regression line moves towards perfection, R2 score move towards 1. And the model performance improves.
# higher R2 score is better
from sklearn.metrics import r2_score
```

```python
#the parameters passed predicted values and actual values(test values)
score = r2_score(reg_pred,y_test)
```
